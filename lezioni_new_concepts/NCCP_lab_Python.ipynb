## 1. Basics of computer programming

- language syntax
- variables
- operations between variables
- control statements
- functions

The *syntax* of a programming language -- similarly to the notion of syntax of a natural language -- defines the set of rules that must be followed to construct a program. If your code complies with the expected syntax, the Python interpreter will "understand" it and thus execute it. Otherwise, a syntax error will be raised.

```python
syntax = 1
if syntax == 1:
  print('I can understand this code!')
```

A variable is a "container" of values, such as numbers or text. The content of a variable might *change* after its initial definition.

```python
name = "Mario"
surname = "Rossi"
birth_year = 1999
current_year = 2025
is_year = True
```

We can also display the content of a variable using the function `print`.

```python
print(birth_year)
print(name)
```

Variables can contain different type of values. To show the type of a variable you can use the function `type`. Python language has some native types such as: string (variables that contains letters and characters), integer (variables that contains intiger digits) and boolean (Logical values, i.e. True or False)

```python
print(type(name), type(birth_year),type(is_year))
```

We can act over variables using a variety of operations, which are usually specific depending on the *type* of variable we are manipulating. For example, we can do arithmetic operations on numeric variables, but not on variables that contain strings of characters (text)!

```python
current_year - birth_year
```

```python
name - surname
```

```python
current_year > birth_year
```

These examples have exploited blocks containing several lines of code, which are called *cells*. Running a cell will produce the corresponding output right below the cell.

If we insert the symbol # before a certain line, we *comment* the corresponding instruction, which will thus be skipped (not executed).

A *program* is made by a sequence of instructions, which are normally executed *from top to bottom*. However, programming languages have specific mechanisms (instructions) called *control statements* that allow to change the execution flow. The most used control statements allows to repeat a block of instructions several times ("loops") or to execute it only if certain conditions are met ("if-then-else").


```python
for passed_years in [2, 3, 4]:  # for each value between 2 and 4, repeat the indented instruction block
  new_year = current_year + passed_years
  print("In", passed_years, "years we will be in", new_year)
```

```python
name = "Sara"

if name == "Mario":  # if the condition is true, execute the following block
  print("Hi Mario!")
else:  # otherwise, execute this one
  print("Hey, I don't know your name!")
```

A portion of your code can be separated from the rest of the program by creating a *function*. A user-defined function has a specific name (that we decide), which allows to *call* it (execute it) when needed.

Python has a set of *built-in* functions, whose names cannot (normally) be used by user-defined functions. The set of built-in functions can be further expanded by *importing* external libraries and modules.

Functions can receive as input some *arguments*, which are variables that can be used in the portion of code of the function.

Functions can give as output some values using the keyword `return`.

```python
def say_hello(input_name):  # input_name is the argument of the function
  print("Hi", input_name, ":)")

def compute_age(birth_year, current_year):
  return current_year - birth_year  # this function returns the computed age
```

```python
say_hello(name)
```

```python
age = compute_age(birth_year, current_year) # the value returned by the function can be stored in a variable
print(age)
```

## 2. Basics of linear algebra with Python

- variables: vectors and matrices
- operations: scalar product, matrix product

A vector is a variable that contains several (ordered) numerical values. Vectors have a *shape*, which specifies how many values are contained, and along with direction the vector develops (horizontal/row or vertical/column). We can print the shape of a vector using the attribute `shape`.

A vector can be multiplied by a scalar number. This operation does not change its shape.

**Vectors and matrices are crucial for scientific computing as well as for implementing neural networks!**

```python
ones_row = np.ones(shape=(1,3))  # create a row vector with three ones
ones_row
```

```python
ones_row * 2
```

```python
zeros_column = np.zeros(shape=(3,1))  # create a column vector with three zeros
zeros_column
```

```python
zeros_column + 3
```

```python
zeros_column.shape
```

A matrix is a vector of two (or more) dimensions. In general, a 2D matrix can have $n$ rows and $m$ columns. Similarly to vectors, also matrices can be multiplied by a scalar number: the result is a matrix of the same shape, where each element is multiplied by the scalar.

Matrices can also be multiplied by a vector o by other matrices using the matrix product. If `A` and `B` are matrices of shape `(n,k)` and `(l,m)`, their matrix product is possible only if `k == l`. The resulting matrix will have shape `(n,m)`, that is, it will have `n` rows and `m` columns.

```python
W = np.random.random_sample(size=(3,4))  # creates a matrix with 3 rows and 4 columns, with random numbers between 0 e 1
W
```

```python
W * 2  # simple product
```

```python
ones_row @ W  # matrix product
```

## 3. Data visualization

The library `matplotlib` allows to create common graphs such as histograms, function plots, bar plots and scatterplots.

```python
import matplotlib.pyplot as plt
```

Let's use `numpy` to sample a set of values from the standard Normal distribution (mean = 0 and std = 1) and then let's visualize the histogram of the data.

```python
random_sample = np.random.standard_normal(size=1000)
# we generate 1000 random values
_ = plt.hist(random_sample)
```

Let's visualize the step function, which returns value $0$ if the input is negative, and $1$ otherwise.

```python
def step(x):
  if x < 0:
    return 0
  else:
    return 1
```

```python
x_coordinates = np.linspace(-3, 3, 100)  # take 100 equally-spaced points between -3 and 3
y_coordinates = [step(c) for c in x_coordinates]  # for each point, compute the output value of the step function

__ = plt.plot(x_coordinates, y_coordinates)
# crate a plot with a line connecting all points
```

Let's now plot the sigmoid function, whose equation is f(x) = {1 \over {1 + e^{-x}}}.$  

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
x_coordinates = np.linspace(-10, 10, 100)
y_coordinates = [sigmoid(c) for c in x_coordinates]

__ = plt.plot(x_coordinates, y_coordinates)
```

## 8) Pandas ‚Äî tiny dataset demo

We'll create a tiny in-memory table (DataFrame), do filtering and a groupby, and plot a simple line.


```python
import pandas as pd
import matplotlib.pyplot as plt  # make sure matplotlib is available

# Create a tiny dataset in memory (no file needed)
df = pd.DataFrame({
    "student": ["A", "B", "C", "D", "E"],
    "quiz":    [1,     1,    1,    2,    2],
    "score":   [7,     8,    6,    9,   10]
})
print(df.head())
print(df.info())

# Filtering and groupby
print(df[df["score"] >= 8])
print(df.groupby("quiz")["score"].mean())

# Plot scores (index on x-axis)
plt.plot(df["score"])
plt.title("Quiz scores by row index")
plt.xlabel("Row index")
plt.ylabel("Score")
plt.show()
```
